# Split the Data

```python
x= np.array(df.iloc[:,[0,1,2,3]].values)
y= np.array(df.iloc[:,4].values)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# Feature Scaling - Normalization (MinMax Scaler) & Standardization (StandardScaler)

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Choosing the Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

# Training the Model

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
final_df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
final_df.head()
```

```python
sns.heatmap(final_df.corr(),annot=True,color='Blue')
```

# Performance Metrics

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
